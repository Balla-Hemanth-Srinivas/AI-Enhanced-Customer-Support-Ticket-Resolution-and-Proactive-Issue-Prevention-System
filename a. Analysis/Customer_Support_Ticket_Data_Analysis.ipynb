```python
# prompt: correlation between each tags(i.e. tag_1, tag_2, tag_3, ........) and priority
from sklearn.preprocessing import LabelEncoder

# Find all tag columns dynamically
tag_columns = [col for col in dataset.columns if 'tag' in col.lower()]

# Create a LabelEncoder object
le_priority = LabelEncoder()

# Fit and transform the 'priority' column
dataset['priority_encoded'] = le_priority.fit_transform(dataset['priority'])

# Iterate through tag columns and calculate correlations
for tag_col in tag_columns:
    # Handle missing values by filling with a placeholder or removing rows
    # Here, we'll fill with 'Unknown' - adjust as needed
    dataset[tag_col] = dataset[tag_col].fillna('Unknown')

    # Create a LabelEncoder for each tag column
    le_tag = LabelEncoder()

    # Fit and transform the current tag column
    dataset[tag_col + '_encoded'] = le_tag.fit_transform(dataset[tag_col])

    # Calculate the correlation
    correlation = dataset[tag_col + '_encoded'].corr(dataset['priority_encoded'])

    print(f"The correlation between '{tag_col}' and 'priority' is: {correlation}")
```

```python
# prompt: find the correlation with priority to all the columns except tags (use percentage to represent the correlation)

from sklearn.preprocessing import LabelEncoder

# Find all tag columns dynamically
tag_columns = [col for col in dataset.columns if 'tag' in col.lower()]

# Create a LabelEncoder object for priority
le_priority = LabelEncoder()
dataset['priority_encoded'] = le_priority.fit_transform(dataset['priority'])

# Iterate through columns (excluding tags and priority) and calculate correlations
for col in dataset.columns:
    if col not in tag_columns and col != 'priority' and col != 'priority_encoded':
        # Handle non-numeric columns using Label Encoding
        if not pd.api.types.is_numeric_dtype(dataset[col]):
            le_col = LabelEncoder()
            dataset[col + '_encoded'] = le_col.fit_transform(dataset[col].fillna('Unknown'))  # Handle missing values
            correlation = dataset[col + '_encoded'].corr(dataset['priority_encoded'])
            print(f"The correlation between '{col}' and 'priority' is: {correlation:.2%}")
        else:
            correlation = dataset[col].corr(dataset['priority_encoded'])
            print(f"The correlation between '{col}' and 'priority' is: {correlation:.2%}")
```

```python
# most repeating word in 'subject' to 'priority' wise

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from collections import Counter
from pprint import pprint
def most_frequent_words_by_priority(df, column_name='subject', priority_column='priority', top_n=5):
    """
    Finds the most frequent words in a specified column, grouped by priority,
    excluding common words and words enclosed in <>.

    Args:
        df: The input DataFrame.
        column_name: The name of the column to analyze (default: 'subject').
        priority_column: The name of the priority column (default: 'priority').
        top_n: The number of most frequent words to return for each priority (default: 3).

    Returns:
        A dictionary where keys are priorities and values are lists of tuples,
        each tuple containing a word and its frequency.
    """
    english_stopwords = set(stopwords.words('english'))
    french_stopwords = set(stopwords.words('french'))
    german_stopwords = set(stopwords.words('german'))
    spanish_stopwords = set(stopwords.words('spanish'))
    portuguese_stopwords = set(stopwords.words('portuguese'))

    # Combine all stop words into a single set
    stop_words = english_stopwords | french_stopwords | german_stopwords | spanish_stopwords | portuguese_stopwords

    # Regex pattern to identify words enclosed in <>
    exclude_pattern = re.compile(r'<[^>]+>')

    results = {}
    for priority in df[priority_column].unique():
        subset = df[df[priority_column] == priority]
        words = []
        for text in subset[column_name].dropna():  # Handle potential missing values
            filtered_words = [
                word for word in re.findall(r'\b\w+\b', text.lower())
                if word not in stop_words and not exclude_pattern.match(word)
            ]
            words.extend(filtered_words)
        word_counts = Counter(words)
        most_common = word_counts.most_common(top_n)
        results[priority] = most_common
    return results

# Example usage (assuming 'dataset' DataFrame is already loaded):
most_frequent = most_frequent_words_by_priority(dataset)
print("The most frequent words in 'subject' according to the 'priority' :- ")
print('----------------------------------------------------------------------')
pprint(most_frequent)
```
